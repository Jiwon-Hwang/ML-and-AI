#### 클래스와 인스턴스
* 클래스 : 붕어빵 기계  
* 인스턴스(객체) : 붕어빵
> ex. 클래스를 사용하면 계산기 대수가 늘어나더라도 객체를 생성하기만 하면 무한히 계산기의 수를 늘릴 수 있다.

```python
class Person:
    def __init__(self, name, age): # 처음에 이름과 나이 입력받을 것
        self.name = name
        self.age = age
    
    def sayHello(self):
        print("Hello 나의 이름은 " + self.name)

p1 = Person("John", 36)
p1.sayHello() # p1 객체가 John의 이름과 나이를 계속 저장해둘 수 있음! (다른 객체와 동시에)

p2 = Person("Jiwon", 25)
p2.sayHello()
p1.sayHello()
```

#### 시그모이드 함수 그리기

```python
import math

X = np.linspace(-10, 10, 100) # num : 총 100개, 10포함
Y = 1/(1 + np.exp(-X)) # 자연상수 e의 -x승

plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
plt.show()
```

#### 시그모이드 함수 미분값

```python
def sigmoid(x):
    s = 1/(1+np.exp(-x))
    ds = s*(1-s) # 시그모이드 함수 1차 미분 함수
    return s, ds

X = np.linspace(-10, 10, 100) 
Y1, Y2 = sigmoid(X) # return : s와 ds
  
plt.plot(X, Y1, X, Y2) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X), Sigmoid'(X)") 
plt.show() 
```

> 중지어 리스트 : wordcloud 모듈의 STOPWORDS에 정리되어 있음. **set()** 형태.   
**합집합(union())**으로 추가 가능

```python
import wikipedia
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# 위키백과 사전의 컨텐츠 제목을 명시
wiki = wikipedia.page('korea') # 'korea'라는 위키백과 사전의 페이지 가져오기

# 이 페이지의 텍스트 컨텐츠를 추출
text = wiki.content # 페이지에서 text 컨텐츠만 추출
s_words = STOPWORDS.union({'one', 'using', 'first', 'two', 'make', 'use', 'south', 'north', 'korean', 'world', 'country'})

# 페이지에서 추출한 텍스트로 워드 클라우드를 생성 => image!
wordcloud = WordCloud(width=2000, height=1500, stopwords=s_words).generate(text) #중지어

plt.figure(figsize=(40, 30))
plt.imshow(wordcloud) # 이미지 show
plt.show() # 그래프 띄우기
```

#### 선형 회귀 분석 => "사이킷런" 라이브러리로 구현
선형 회귀 : 임의의 변수 x(데이터의 특징)와 이에 따른 변수 y와의 상관관계를 모델링하는 기법

```python
from sklearn import linear_model 

regr = linear_model.LinearRegression() # 선형회귀 모델 생성!

x = [[164] ,[179], [162], [170]] # 다중회귀에도 사용하도록 함
y = [53, 63, 55, 59] # y = f(x)의 결과
regr.fit(x, y) # x와 y를 맞추기 => ***fit() : 모델 생성 함수 (모델 생성기)
# 최적화 시킴
```

### 주의 : 선형회귀의 입력(X)은 반드시 [[다차원]] 배열 사용!
=> [164, ...] 과 같은 1차원 배열 값 사용하면 오류 발생

```python
coef = regr.coef_ # 직선의 기울기 (coefficient : 계수)
intercept = regr.intercept_ # 절편
score = regr.score(x, y) # 얼마나 잘 예측하는가 (예측 정확도)

print('기울기(coef) :' , coef)
print('절편(intercept) :', intercept)
print('학습된 직선이 데이터를 얼마나 잘 따르는지(score) :', score)
```

```python
input_data = [[180], [185]]

result = regr.predict(input_data) # 입력에 따른 학생들의 몸무게 추정, 예측
print(result)
```

```python
# 입력된 학습 데이터인 x와 y를 산포도로 그리기
plt.scatter(x, y, color='black')

# 학습 데이터를 다시 입력으로 하여 예측값 계산
y_pred = regr.predict(x)
print(y_pred)

# 학습데이터와 예측값을 선그래프로 그리기
plt.plot(x, y_pred, color='blue', linewidth=3)
plt.show()
```

#### KNN알고리즘과 Iris 종류 예측
> * cf. KNN : 입력값과 가장 유사한 k개를 찾아서 가장 많은 수를 차지하는 label을 출력
> * But, KNN : 학습보다 테스트가 더 오래걸려서 분류에 잘 쓰지 않음

```python
# (80:20)으로 분할 => 80은 학습용, 20은 테스트용
from sklearn.datasets import load_iris # (from 모듈 import 함수) => "namespace" 생략 가능!
from sklearn.model_selection import train_test_split

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

num_neigh = 3 # k개수 설정
knn = KNeighborsClassifier(n_neighbors = num_neigh) # knn 모델 생성
knn.fit(X_train, y_train) # knn 모델 학습
y_pred = knn.predict(X_test) # 예측(테스트)
scores = metrics.accuracy_score(y_test, y_pred) # 모델 예측 정확도 평가 (y_test와 비교)
```

```python
print('k가 {0:d}일때 정확도 : {1:.3f}'.format(num_neigh, scores))
```
