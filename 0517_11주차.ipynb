### Data로 차트(그래프 등) 그리기
df.plot(x=열, y=열) 
> plt.plot과 조금 다름

```python
# 데이터 정의
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'name':['Kim','Lee','Park','Choi','Hong','Chung','Jang'],
    'age':[22,26,78,17,46,32,21],
    'city':['Seoul','Busan','Seoul','Busan','Seoul','Daejun','Daejun'],
    'children':[2,3,0,1,3,4,3],
    'pets':[0,1,0,2,2,0,3]
})

# 그래프 그리기 (각각 사람들이 키우는 애완동물 수)
df.plot(kind='line', x='name', y='pets', color='red') # kind='line' : 디폴트
plt.show()
```

#### 중첩 그래프 그리기
ax (하나의 축) 생성 후, df.plot()으로 그릴때 인자로 두 그래프에 모두 ax=ax 해줘야 함!

```python
ax = plt.gca() # axes (축, 그래프)가 없을 경우 생성

df.plot(kind='line',x='name',y='children',ax=ax) # 같은 축에 생성
df.plot(kind='line',x='name',y='pets', color='red', ax=ax) # 같은 축에 생성

plt.show()
```

#### 그룹핑하여 그리기
df.groupby()[]

```python
# 도시명 기준으로 사람 이름을 그룹화해서 nunique : 사람 이름의 개수를 구하기
df.groupby('city')['name'].nunique().plot(kind='bar')
plt.show()
```

```python
df.groupby('city')['name']
```

```python
df.groupby('city').nunique().plot(kind='bar') # 모든 열들을 도시명을 기준으로 그룹화
plt.show()
```

#### 피벗 테이블에서 인덱스를 사용하여 그룹화

#### 다중 인덱스로 피봇

```python
table = pd.pivot_table(titanic, index=['Sex','Pclass'])
table
```

#### 특징별로 다른 집계 함수 적용

```python
table = pd.pivot_table(titanic ,index=['Sex','Pclass'],aggfunc={'Age':np.mean,'Survived':np.sum})
table
```

#### value 매개 변수를 사용해 특정 데이터만 집계

```python
table=pd.pivot_table(titanic, index=['Sex', 'Pclass'], values=['Survived'], aggfunc=np.mean)
table
```

```python
table.plot(kind="bar")
```

```python
table = pd.pivot_table(titanic ,index=['Sex'],columns=['Pclass'],values=['Survived'],aggfunc=np.sum)
table.plot(kind="bar")
```

```python
table # 성별 기준으로(index : 행), 열에는 등급이 오도록 그룹화. 그룹화할 대상은 생존자 수 합
```

#### merge()

```python
df1 = pd.DataFrame({'employee': ['Kim', 'Lee', 'Park', 'Choi'],
                    'department': ['Accounting', 'Engineering', 'HR', 'Engineering']})
df2 = pd.DataFrame({'employee': ['Kim', 'Lee', 'Park', 'Choi'],
                    'age': [27, 34, 26, 29]})
df3 = pd.merge(df1, df2)

df3
```

### 결손값 삭제하기!
> dropna() : NaN으로 표현된 결손값(누락값)을 탐지하고 수정, 삭제하는 함수

```python
import pandas as pd
df = pd.read_csv('countries1.csv', index_col=0) # 맨왼쪽 인덱스가 될 열 지정(code)
df
```

```python
# 결손값 모두 제거
df.dropna(how="all") # 행의 모든 값들이 NaN일때만 그 행 전부 제거
```

```python
df.dropna() # 하나라도 있으면 그 행 전부 제거
```

```python
df_0 = df.fillna(0) # 결손값 0으로 보정
df_0
```

```python
df.fillna(df.mean()['area']) # area 열의 평균값으로 결손값 보정
```

```python
df.mean() # 평균을 낼 수 있는 열들(숫자로 이루어진 열)은 평균 내기
```
